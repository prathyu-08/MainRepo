```python
def apply_operation(func, x, y):
    return func(x, y)

def add(a, b):
    return a+b

def sub(a, b):
    return a-b

print(apply_operation(add, 6, 4))
print(apply_operation(sub, 6, 4))
```

```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]
sum_of_numbers = reduce(lambda x, y: x+y, numbers)
print(sum_of_numbers)
```

```python
numbers = [1, 3, 4, 5, 6, 7]
even_numbers = list(filter(lambda x: x% 2 ==0, numbers))
print(even_numbers)
```
